```python
import numpy as np
# Set a vector of discretized cake sizes
W_lb = 1e-2
W_ub = 1.0
N = 100
W_vec = np.linspace(W_lb, W_ub, N)
print(W_vec)
```

```python
# 5.10
```

```python
# Choose a utility function u(c) = log(c)
def utility(c):
    util = np.zeros_like(c)
    util = np.log(c)
    return util
beta = 0.9
```

```python
v_init = utility(W_vec)
V_prime = np.zeros(N).reshape(N, 1)
```

```python
# Create utility matrix
c_mat = (np.tile(W_vec.reshape((N, 1)), (1, N)) - np.tile(W_vec.reshape((1, N)), (N, 1)))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7
u_mat = utility(c_mat)
VF = u_mat + beta * V_prime
V_new = VF.max(axis = 1)
ind = np.argmax(u_mat + beta * V_prime, axis = 1)
W_vec[ind]
```

```python
plt.plot(W_vec, W_vec[ind])
plt.title("Policy Function")
```

```python
plt.plot(V_new[1:])
plt.title("Value Function")
```

```python
# 5.14
```

```python
N = 100
c_mat = (np.tile(W_vec.reshape((N, 1)), (1, N)) - 
         np.tile(W_vec.reshape((1, N)), (N, 1)))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7
u_mat = utility(c_mat)
maxiters = 500
toler = 1e-10
dist = 10.0
VF_iter = 0
V_init = utility(W_vec)
while dist > toler and VF_iter < maxiters:
    VF_iter += 1
    # One contraction mapping
    V_prime = np.tile(V_init.reshape((1, N)), (N, 1))
    V_prime[~c_pos] = -9e+4
    VF = u_mat + beta * V_prime
    V_new = (u_mat + beta * V_prime).max(axis = 1)
    ind = np.argmax(VF, axis = 1)
    dist = ((V_new - V_init) ** 2).sum()
    print('Iter=', VF_iter, ', distance =', dist)
    V_init = V_new
print("It takes", VF_iter, "iterations to converge")
print("Does it converge to the fixed point?", np.array_equal(V_init, V_new))
```

```python
sigma = 0.5
mu = 4 * sigma
M = 7
epsilon = np.linspace(mu - 3 * sigma, mu + 3 * sigma, M)
epsilon
```

```python
from scipy.stats import norm
pr = lambda x: norm(loc = mu, scale = sigma).pdf(x)
pdf = pr(epsilon)
pdf
```

```python
plt.plot(epsilon, pdf)
```

```python
# 5.17
```

```python
c_mat = (np.tile(W_vec.reshape((N, 1)), (1, N)) - np.tile(W_vec.reshape((1, N)), (N, 1)))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7
u_mat = utility(c_mat)
cube = np.array([u_mat*e for e in epsilon])
V_init = np.zeros((N, M))
ev = V_init @ pdf.reshape((M, 1))
ev_mat = np.tile(ev.reshape((1, N)), (N, 1))
ev_mat[~c_pos] = -9e+5
ev_cube = np.array([ev_mat for e in range(M)])
vt = cube + beta * ev_cube
V_new = np.zeros((N,M))
W_new = np.zeros((N, M))
for n in range(N):
    VF = vt[:, n, :]
    V_new[n] = VF.max(axis = 1)
    ind = np.argmax(VF, axis = 1)
    W_new[n] = W_vec[ind]
```

```python
plt.plot(W_vec, np.average(W_new, axis = 1))
plt.title("Policy Function")
```

```python
plt.plot(np.average(V_new, axis =1))
plt.title("Value Function")
```
